```python
## Creating a column for Total credit score of a person
pd.set_option('display.max_columns',None)
```

```python
df['Total_Credit_Score'] = df['Total_Good_Debt'] - df['Total_Bad_Debt']
```

```python
df.head(3)
```

```python
## making a copy of the df for ml
```

```python
ml = df.copy()
```

```python
ml.head()
```

```python
## Droping unwanted columns for prediction
```

```python
# ml.drop(['Applicant_ID','Total_Income','Income_Type','Education_Type','Family_Status','Housing_Type','Owned_Mobile_Phone','Owned_Work_Phone','Owned_Phone','Owned_Email','Total_Family_Members'],axis=1,inplace=True)
```

```python
ml.head()
```

```python
## M - 1 ; F - 0
## Highest Education - Max number ; lowest education - min number
```

```python
from sklearn.preprocessing import LabelEncoder, OrdinalEncoder
```

```python
ml['Applicant_Gender'].unique()
```

```python
ml['Applicant_Gender'] = ml['Applicant_Gender'].map({'M      ':1,'F      ':0})
```

```python
ml['Education_Type'] = ml['Education_Type'].str.strip()
```

```python
ml['Education_Type'].unique()
```

```python
## Ordinal encoding Education_Type
```

```python
oe_edu = OrdinalEncoder(categories=[['Lower secondary','Secondary / secondary special','Incomplete higher','Higher education','Academic degree']])
```

```python
ml['Education_Type'] = oe_edu.fit_transform(ml[['Education_Type']])
```

```python
## Label Encoding Income_Type
```

```python
le_IT = LabelEncoder()
```

```python
ml['Income_Type'] = le_IT.fit_transform(ml['Income_Type'])
```

```python
## Label Encoding Income_Type 'Family_Status'
```

```python
le_FS = LabelEncoder()
```

```python
ml['Family_Status'] = le_FS.fit_transform(ml['Family_Status'])
```

```python
## Label Encoding Income_Type 'Housing_Type'
```

```python
le_HT = LabelEncoder()
ml['Housing_Type'] = le_FS.fit_transform(ml['Housing_Type'])
```

```python
## Label Encoding Income_Type 'Job_Title'
```

```python
le_JT = LabelEncoder()
ml['Job_Title'] = le_FS.fit_transform(ml['Job_Title'])
```

```python
tl=ml.copy()
```

```python
tl.drop('Total_Credit_Score',axis=1,inplace=True)
```

```python
tl.head()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier, BaggingClassifier
from catboost import CatBoostClassifier
from lightgbm import LGBMClassifier
from xgboost import XGBClassifier, XGBRFClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
# x = tl.drop('Status',axis=1)
# y = tl[['Status']]

x = ml.drop('Status',axis=1)
y = ml[['Status']]
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
lr = LogisticRegression()
```

```python
lr.fit(xtrain,ytrain) ## to train using xtrain and ytrain
```

```python
lr_preds = lr.predict(xtest)
```

```python
lr_preds
```

```python
len(lr_preds)
```

```python
xtest.shape
```

```python
lr_accuracy = round(accuracy_score(ytest,lr_preds)*100,2)
```

```python
lr_accuracy
```

```python
lr_conf_matrix = confusion_matrix(ytest,lr_preds)
```

```python
lr_conf_matrix
```

### my positive class is 1 and negative class is 0 here
### True negative = 21, False Positive = 5, False Negative = 1, True Positive = 4999
### Since this is a credit card approval problem, reducing false positives (5) is more important than reducing false negatives (1)
### Approving a credit card for a high-risk customer can result in financial loss, while rejecting a low-risk customer mainly leads to missed opportunity.
### So the model evaluation focuses more on Recall rather than overall accuracy

```python
lr_class_report = classification_report(ytest,lr_preds)
```

```python
print(lr_class_report)
```

```python
pd.DataFrame(lr_preds).value_counts()
```

```python
lr_final = xtest.copy()
lr_final['LR_PREDS'] = lr_preds
lr_final['ACTUALS'] = ytest
lr_final.head()
```

```python
lr_final[(lr_final['ACTUALS'] == 1) & (lr_final['LR_PREDS'] == 0)]
```

### Logistic Regression struggled to capture non linear relationships in the data, So going to use of tree based models

```python
print('\033[1mLOGISTIC REGRESSION\033[0m','\n')
print('\033[1mAccuracy Score: \033[0m', lr_accuracy, '\n')
print('\033[1mConfusion Matix: \033[0m','\n', lr_conf_matrix, '\n')
print('\033[1mClassification Report: \033[0m','\n', lr_class_report)
```

## Decision Tree

```python
dt = DecisionTreeClassifier()
dt.fit(xtrain, ytrain)
dt_preds = dt.predict(xtest)
dt_acc = accuracy_score(ytest, dt_preds)
dt_conf = confusion_matrix(ytest, dt_preds)
dt_clas = classification_report(ytest, dt_preds)
print('\033[1mDECISION TREE\033[0m','\n')
print('\033[1mAccuracy Score: \033[0m', dt_acc, '\n')
print('\033[1mConfusion Matix: \033[0m','\n', dt_conf, '\n')
print('\033[1mClassification Report: \033[0m','\n', dt_clas)
```
